```python
import pandas as pd
import nltk
from nltk import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
8#se importan los datos en foramto csv
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')
data
```

```python
data["reviewText"].values.tolist()#se organiza el dataframe como una lista de datos
```

```python
tokenized_data = []#se crea una lista vacia
for text in data["reviewText"].values.tolist():#se realiza un ciclo for
    tokens = word_tokenize(text)#con la herramienta de word_tokenize se dividen las oraciones
    tokenized_data.append(tokens)#se añaden los tokens y se ponen en listas

for tokens in tokenized_data:# se realiza un ciclo for
    print(tokens)#se imprimen los tokens
```

```python
stop_words = stopwords.words('english')#se descartan las palabras vacias, es decir las stopwords
print(stop_words)
```

```python
words = [w.lower() for w in tokens]#se ponen todas las palabras en minuscula
print(words)
```

```python
words = [w for w in words if not w in stop_words and w.isalpha()]#se dejan las palabras importantes, y se quitan todas las demas que sean repetidas o en algunos casos prreposiciones y concetores de oraciones
print(words)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

#Se perzonalisa para el tipo de parametros que queremos ver en nuestra nube
textoWC = ' '.join(words)
wordcloud = WordCloud(
    width = 600,
    height = 600,
    background_color = 'linen',
    stopwords = STOPWORDS,
    relative_scaling = 0,
    min_font_size = 10
).generate(textoWC)
plt.figure(figsize = (7,7), facecolor = None)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
```
